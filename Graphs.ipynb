```python
import numpy as np
import matplotlib.pyplot as plt

# Данные: температуры и ln(η)
T_steklo = np.array([1/293, 1/301, 1/309, 1/317, 1/325, 1/333])
ln_eta_steklo = np.array([6.888, 6.345, 5.756, 5.273, 4.869, 4.504])

T_metall = np.array([1/293, 1/301, 1/309, 1/317, 1/325, 1/333])
ln_eta_metall = np.array([6.997, 6.368, 5.681, 5.316, 4.761, 4.451])

def mnk_line(x, y):
    """Рассчет коэффициентов прямой y = kx + b методом наименьших квадратов"""
    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    # Считаем коэффициенты k и b
    k = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
    b = y_mean - k * x_mean

    return k, b

# Получаем параметры прямых
k_steklo, b_steklo = mnk_line(T_steklo, ln_eta_steklo)
k_metall, b_metall = mnk_line(T_metall, ln_eta_metall)

# Генерируем точки для прямых по этим коэффициентам
T_range = np.linspace(1/333, 1/293, 100)
ln_eta_steklo_fit = k_steklo * T_range + b_steklo
ln_eta_metall_fit = k_metall * T_range + b_metall

# Строим график
plt.figure(figsize=(10, 6))

# Точки
plt.scatter(T_steklo, ln_eta_steklo, color='blue', label='Стекло (данные)')
plt.scatter(T_metall, ln_eta_metall, color='red', label='Металл (данные)')

# Линии МНК
plt.plot(T_range, ln_eta_steklo_fit, color='blue', linestyle='--', label=f'МНК Стекло: ln(η) = {k_steklo:.4f}T + {b_steklo:.2f}')
plt.plot(T_range, ln_eta_metall_fit, color='red', linestyle='--', label=f'МНК Металл: ln(η) = {k_metall:.4f}T + {b_metall:.2f}')

plt.xlabel('Температура, K')
plt.ylabel('ln(η)')
plt.title('График ln(η) от температуры с линиями МНК')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
x1 = np.array([2.72, 10.59, 26.28, 51.77, 81.14, 121.84, 165.43, 223.85, 280.55, 297.95])
y = np.array([23.5, 23.52, 23.6, 23.71, 23.85, 24.06, 24.25, 24.52, 24.79, 24.87])

# Твои значения i и u
i = np.array([10.75, 21.22, 33.37, 46.72, 58.33, 71.17, 82.59, 95.54, 106.39, 106.46])
u = np.array([252.7, 499.1, 787.5, 1108, 1391, 1712, 2003, 2343, 2637, 2722])

# Погрешности (относительные)
rel_error = ((((0.012 * u * 0.001 * 0.01 + 0.00002) / (u * 0.001)) ** 2 +
              ((0.15 * i * 0.001 * 0.01 + 0.00002) / (i * 0.001)) ** 2) ** 0.5)

# Абсолютные погрешности для x1 и y
sigma_x1 = rel_error * x1
sigma_y = rel_error * y

# Весовые коэффициенты: используем ошибки y
wi = 1 / sigma_y**2

# Средние значения по весам
xw_mean = np.sum(wi * x1) / np.sum(wi)
yw_mean = np.sum(wi * y) / np.sum(wi)

# Угловой коэффициент k (метод МНК с весами)
numerator = np.sum(wi * (x1 - xw_mean) * (y - yw_mean))
denominator = np.sum(wi * (x1 - xw_mean)**2)
k = numerator / denominator

# Погрешность углового коэффициента
sigma_k = np.sqrt(1 / denominator)

# Свободный член
b = yw_mean - k * xw_mean

# Вывод результатов
print(f"Угловой коэффициент k = {k:.5f}")
print(f"Погрешность углового коэффициента σ_k = {sigma_k:.5f}")
print(f"Свободный член b = {b:.5f}")

# График
x1_range = np.linspace(min(x1), max(x1), 100)
y_fit = k * x1_range + b

plt.errorbar(x1, y, xerr=sigma_x1, yerr=sigma_y, fmt='o', label='Данные с погрешностями')
plt.plot(x1_range, y_fit, color='red', label=f'МНК: y = {k:.4f} * x + {b:.2f}')

plt.xlabel('x1')
plt.ylabel('y')
plt.title('Линейная аппроксимация МНК с ошибками')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
x1 = np.array([2.72, 10.59, 26.28, 51.77, 81.14, 121.84, 165.43, 223.85, 280.55, 297.95])
y = np.array([23.5, 23.52, 23.6, 23.71, 23.85, 24.06, 24.25, 24.52, 24.79, 24.87])
i = np.array([10.75, 21.22, 33.37, 46.72, 58.33, 71.17, 82.59, 95.54, 106.39, 106.46])
u = np.array([252.7, 499.1, 787.5, 1108, 1391, 1712, 2003, 2343, 2637, 2722])

# Погрешности (относительные), затем абсолютные
t = y * ((((0.012 * u * 0.001 * 0.01 + 0.00002) / (u * 0.001)) ** 2 +
          ((0.15 * i * 0.001 * 0.01 + 0.00002) / (i * 0.001)) ** 2) ** 0.5)

# Построение прямой по polyfit
a6, b = np.polyfit(x1, y, 1)

# Рассчитываем погрешность наклона вручную
n = len(x1)
y_fit = a6 * x1 + b

# Остатки
residuals = y - y_fit

# Сумма квадратов остатков
S_res = np.sum(residuals**2)

# Сумма квадратов по x относительно среднего
S_xx = np.sum((x1 - np.mean(x1))**2)

# Погрешность наклона
sigma_k = np.sqrt(S_res / (n - 2)) / np.sqrt(S_xx)

# Выводим результат
print(f"Коэффициент наклона k = {a6:.5f}")
print(f"Свободный член b = {b:.5f}")
print(f"Погрешность углового коэффициента sigma_k = {sigma_k:.5f}")

# График
x_line = np.linspace(min(x1), max(x1), 1000)

# Интерполяция значений y
y_interp = np.interp(x_line, x1, y)

# Интерполяция погрешностей
t_interp = np.interp(x_line, x1, t)

# Закрашивание области погрешностей относительно точек
plt.fill_between(x_line, y_interp - t_interp, y_interp + t_interp, alpha=0.3, color='green', label='Погрешность')

# Построение точек
plt.scatter(x1, y, color='red', s=10, zorder=3)

# Построение линии регрессии
y_line = a6 * x_line + b
plt.plot(x_line, y_line, label=f"y = {a6:.5f}x + {b:.3f}", color='blue')

# Добавление погрешностей
plt.errorbar(x1, y, xerr=t, yerr=t, linestyle='', color='black', capsize=2.5)

# Оформление графика
plt.grid()
plt.xlabel("Q, мВт")
plt.ylabel("Rн, Ом")
plt.title("Зависимость Rн от Q, Т=340.4 K")
plt.legend()

# Сохранение графика
#plt.savefig('C:\\prog\\Зависимость Rн от Q, 67.4.pdf')

plt.show()

# Печать массива погрешностей для проверки
print("Абсолютные погрешности t:", t)
```
